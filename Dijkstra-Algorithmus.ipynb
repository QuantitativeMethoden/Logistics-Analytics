```python
from queue import PriorityQueue
```

```python
# Erstelle eine neue Klasse und definiere ihre Eigenschaften.
class Graph:
    # Definition was die Klasse "Graph" überhaupt ist und aus was sie besteht
    def __init__(self, eckpunkte):
        self.v = eckpunkte
        self.kanten = [[-1 for i in range(eckpunkte)] for j in range(eckpunkte)]
        self.besucht = []
    # Definition wie Kanten hinzugefügt werden    
    def add_kante(self, u, v, weight):
        self.kanten[u][v] = weight
        self.kanten[v][u] = weight
    # Implementierung Dijktra Algorithmus    
    def dijkstra(self, startpunkt):
        D = {v:float('inf') for v in range(self.v)}
        D[startpunkt] = 0
    
        pq = PriorityQueue()
        pq.put((1, startpunkt))
    
        while not pq.empty():
            (dist, aktueller_eckpunkt) = pq.get()
            self.besucht.append(aktueller_eckpunkt)
    
            for nachbar in range(self.v):
                if self.kanten[aktueller_eckpunkt][nachbar] != -1:
                    entfernung = self.kanten[aktueller_eckpunkt][nachbar]
                    if nachbar not in self.besucht:
                        alte_kosten = D[nachbar]
                        neue_kosten = D[aktueller_eckpunkt] + entfernung
                        if neue_kosten < alte_kosten:
                            pq.put((neue_kosten, nachbar))
                            D[nachbar] = neue_kosten
        return D        
```

```python
# Festlegen der Eckpunkte des Graphen
g = Graph(6)

#Festlegen der Verbindungen zwischen den einzelnen Eckpunkte und deren Gewichtung
g.add_kante(0, 1, 0)
g.add_kante(1, 2, 20)
g.add_kante(1, 4, 10)
g.add_kante(2, 3, 20)
g.add_kante(2, 5, 50)
g.add_kante(2, 4, 10)
g.add_kante(5, 3, 20)
g.add_kante(3, 5, 10)
```

```python
# Ausführen Dijkstra und Ausgabe der Entfernungen
D = g.dijkstra(1)
for vertex in range(len(D)):
    if vertex > 0:
        print("Entfernung vom Startpunkt zum Eckpunkt", vertex, "ist", D[vertex])
```
